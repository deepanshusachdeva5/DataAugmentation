```python
dc = 0# dog count
cc = 0#cat count

training_size = 1000
test_size = 500

training_images = []
training_labels = []
test_images = []
test_labels = []

dog_dir_train = 'datasets/train/dog/'
cat_dir_train  = 'datasets/train/cat/'
dog_dir_test = 'datasets/test/dog/'
cat_dir_test = 'datasets/test/cat/'




```

```python
def make_dir(directory):
        if os.path.exists(directory):
            shutil.rmtree(directory)
        os.makedirs(directory)
```

```python
make_dir(dog_dir_train)
make_dir(cat_dir_train)
make_dir(dog_dir_test)
make_dir(cat_dir_test)
```

```python
def getZeros(number):
    if(number > 10 and number < 100):
        return "0"
    if(number < 10):
        return "00"
    else:
        return ""
```

```python
for i,file in enumerate(files_names):
    
    if files_names[i][0] == 'd':
        dc+=1
        image = cv2.imread(mypath+file)
        image = cv2.resize(image, (150, 150), interpolation = cv2.INTER_AREA)
        
        if dc<=training_size:
            training_images.append(image)
            training_labels.append(1)
            zeros = getZeros(dc)
            cv2.imwrite(dog_dir_train + "dog" + str(zeros) + str(dc) + ".jpg", image)
            
        elif dc > training_size  and dc<=test_size+training_size:
            test_images.append(image)
            test_labels.append(1)
            zeros = getZeros(dc-1000)
            cv2.imwrite(dog_dir_test + "dog" + str(zeros) + str(dc-1000) + ".jpg", image)
        
    
    elif files_names[i][0] == 'c':
        cc+=1
        image = cv2.imread(mypath+file)
        image = cv2.resize(image, (150, 150), interpolation = cv2.INTER_AREA)
        
        if cc<=training_size:
            training_images.append(image)
            training_labels.append(0)
            zeros = getZeros(cc)
            cv2.imwrite(cat_dir_train + "cat" + str(zeros) + str(cc) + ".jpg", image)
            
        elif cc > training_size  and cc<=test_size+training_size:
            test_images.append(image)
            test_labels.append(0)
            zeros = getZeros(cc-1000)
            cv2.imwrite(cat_dir_test + "cat" + str(zeros) + str(cc-1000) + ".jpg", image)

        
    if dc == training_size+test_size and cc == training_size+test_size:
        break
    
print("TESTING AND TRAINING IMAGES EXTRACTION COMPLETE!")
```

```python
np.savez('cats_vs_dogs_training_data.npz', np.array(training_images))
np.savez('cats_vs_dogs_training_labels.npz', np.array(training_labels))
np.savez('cats_vs_dogs_test_data.npz', np.array(test_images))
np.savez('cats_vs_dogs_test_labels.npz', np.array(test_labels))
```

```python
def load_data_training_and_test(datasetname):
    
    npzfile = np.load(datasetname + "_training_data.npz")
    train = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_training_labels.npz")
    train_labels = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_test_data.npz")
    test = npzfile['arr_0']
    
    npzfile = np.load(datasetname + "_test_labels.npz")
    test_labels = npzfile['arr_0']

    return (train, train_labels), (test, test_labels)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Dropout, Flatten
```

```python
from keras.layers import Conv2D,MaxPooling2D
```

```python

model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(x_train.shape[1],x_train.shape[2],x_train.shape[3])))
model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))


model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dense(1,activation='sigmoid'))


model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])


print(model.summary())



```

```python
model.fit(x_train,y_train,batch_size=64,epochs=2,validation_data=(x_test,y_test))
```
